```python
import os
import pandas as pd
import pandasql as ps
import numpy as np
from datetime import datetime

adnifiltered=pd.read_excel("/mnt/f/Tirocinio_HDD/Data/ADNI/MERGED_FILES_FINAL.xlsx","MAIN-Filtered")
FDG=pd.read_excel("/mnt/f/Tirocinio_HDD/Data/ADNI/FDG_Analysis_Filtered.xlsx")
FBB=pd.read_excel("/mnt/f/Tirocinio_HDD/Data/ADNI/FBB_Analysis_Filtered.xlsx")
AV45=pd.read_excel("/mnt/f/Tirocinio_HDD/Data/ADNI/AV45_Analysis_Filtered.xlsx")
AV1451=pd.read_excel("/mnt/f/Tirocinio_HDD/Data/ADNI/AV1451_Analysis_Filtered.xlsx")
VOLUMES=pd.read_excel("/mnt/f/Tirocinio_HDD/Data/ADNI/TOTAL_VOLUMES.xlsx")
```

```python
total=adnifiltered.copy()
total=total.drop(['index'],axis=1)
colsVA45_FBB=['A_COMPOSITE_SUVR',
 'A_NORM_CTX_CAUDALMIDDLEFRONTAL_SUVR',
 'A_NORM_CTX_ENTORHINAL_SUVR',
 'A_NORM_AMYGDALA_SUVR',
 'A_NORM_HIPPOCAMPUS_SUVR']

colsV1A451=['T_COMPOSITE_SUVR',
 'T_NORM_CTX_CAUDALMIDDLEFRONTAL_SUVR',
 'T_NORM_CTX_ENTORHINAL_SUVR',
 'T_NORM_AMYGDALA_SUVR',
 'T_NORM_HIPPOCAMPUS_SUVR']


colsVOL=['N_HIPPOCAMPUS_VOLUME', 'N_AMYGDALA_VOLUME','N_CTX_ENTORHINAL_VOLUME','N_CTX_CAUDALMIDDLEFRONTAL_VOLUME']

colsFDG=['F_NORM_MEAN']


cols=[*colsVA45_FBB,*colsV1A451,*colsFDG,*colsVOL,*colsFDG]

tmp=pd.DataFrame(columns=cols)
```

```python
total=total.join(tmp)
```

```python
ch=True
for row in range(0,len(total)): #per ogni riga verifica gli 1/0

    if total.loc[row].AV45 == 1:
        if total.loc[row].FBB == 1 and ch==True:
            ch=False
            #addA(AV45)
        else:
            total.loc[row,'A_COMPOSITE_SUVR'] = AV45[AV45.RID==total.loc[row].RID].COMPOSITE_SUVR.values[0]                                       
            total.loc[row,'A_NORM_CTX_CAUDALMIDDLEFRONTAL_SUVR']  = (AV45[AV45.RID==total.loc[row].RID].AV45_NORM_CTX_LH_CAUDALMIDDLEFRONTAL_SUVR.values[0]+AV45[AV45.RID==total.loc[row].RID].AV45_NORM_CTX_RH_CAUDALMIDDLEFRONTAL_SUVR.values[0])/2                  
            total.loc[row,'A_NORM_CTX_ENTORHINAL_SUVR']  = (AV45[AV45.RID==total.loc[row].RID].AV45_NORM_CTX_LH_ENTORHINAL_SUVR.values[0] +AV45[AV45.RID==total.loc[row].RID].AV45_NORM_CTX_RH_ENTORHINAL_SUVR.values[0])/2                      
            total.loc[row,'A_NORM_AMYGDALA_SUVR']  = (AV45[AV45.RID==total.loc[row].RID].AV45_NORM_LEFT_AMYGDALA_SUVR.values[0]+AV45[AV45.RID==total.loc[row].RID].AV45_NORM_LEFT_AMYGDALA_SUVR.values[0])/2                                  
            total.loc[row,'A_NORM_HIPPOCAMPUS_SUVR']  = (AV45[AV45.RID==total.loc[row].RID].AV45_NORM_LEFT_HIPPOCAMPUS_SUVR.values[0]+AV45[AV45.RID==total.loc[row].RID].AV45_NORM_LEFT_HIPPOCAMPUS_SUVR.values[0])/2                               



        
    if total.loc[row].FBB == 1:
        if total.loc[row].AV45 == 1 and ch==True:
            ch=False
           # addA(FBB)
        else:
            total.loc[row,'A_COMPOSITE_SUVR'] = FBB[FBB.RID==total.loc[row].RID].COMPOSITE.values[0]  
            total.loc[row,'A_NORM_CTX_CAUDALMIDDLEFRONTAL_SUVR']  = (FBB[FBB.RID==total.loc[row].RID].FBB_NORM_CTX_LH_CAUDALMIDDLEFRONTAL_SUVR.values[0] +FBB[FBB.RID==total.loc[row].RID].FBB_NORM_CTX_LH_CAUDALMIDDLEFRONTAL_SUVR.values[0])/2             
            total.loc[row,'A_NORM_CTX_ENTORHINAL_SUVR']  = (FBB[FBB.RID==total.loc[row].RID].FBB_NORM_CTX_LH_ENTORHINAL_SUVR.values[0]   +FBB[FBB.RID==total.loc[row].RID].FBB_NORM_CTX_LH_ENTORHINAL_SUVR.values[0] )/2                    
            total.loc[row,'A_NORM_AMYGDALA_SUVR']  = (FBB[FBB.RID==total.loc[row].RID].FBB_NORM_LEFT_AMYGDALA_SUVR.values[0]  + FBB[FBB.RID==total.loc[row].RID].FBB_NORM_LEFT_AMYGDALA_SUVR.values[0] )/2                          
            total.loc[row,'A_NORM_HIPPOCAMPUS_SUVR']  = (FBB[FBB.RID==total.loc[row].RID].FBB_NORM_LEFT_HIPPOCAMPUS_SUVR.values[0]  +FBB[FBB.RID==total.loc[row].RID].FBB_NORM_LEFT_HIPPOCAMPUS_SUVR.values[0])/2                          


  
            
    if total.loc[row].AV1451 == 1:
  
       
        total.loc[row,'T_COMPOSITE_SUVR'] = AV1451[AV1451.RID==total.loc[row].RID].COMPOSITE.values[0]                                       
        total.loc[row,'T_NORM_CTX_CAUDALMIDDLEFRONTAL_SUVR']  = (AV1451[AV1451.RID==total.loc[row].RID].AV1451_NORM_CTX_LH_CAUDALMIDDLEFRONTAL_SUVR.values[0] + AV1451[AV1451.RID==total.loc[row].RID].AV1451_NORM_CTX_LH_CAUDALMIDDLEFRONTAL_SUVR.values[0] )/2                 
        total.loc[row,'T_NORM_CTX_ENTORHINAL_SUVR']  = (AV1451[AV1451.RID==total.loc[row].RID].AV1451_NORM_CTX_LH_ENTORHINAL_SUVR.values[0]  + AV1451[AV1451.RID==total.loc[row].RID].AV1451_NORM_CTX_LH_ENTORHINAL_SUVR.values[0])/2                     
        total.loc[row,'T_NORM_AMYGDALA_SUVR']  = (AV1451[AV1451.RID==total.loc[row].RID].AV1451_NORM_LEFT_AMYGDALA_SUVR.values[0]  +AV1451[AV1451.RID==total.loc[row].RID].AV1451_NORM_LEFT_AMYGDALA_SUVR.values[0] )/2                             
        total.loc[row,'T_NORM_HIPPOCAMPUS_SUVR']  = (AV1451[AV1451.RID==total.loc[row].RID].AV1451_NORM_LEFT_HIPPOCAMPUS_SUVR.values[0]   +AV1451[AV1451.RID==total.loc[row].RID].AV1451_NORM_LEFT_HIPPOCAMPUS_SUVR.values[0] )/2                                                  
        
    if total.loc[row].FDG == 1:
        total.loc[row,'F_NORM_MEAN']=FDG[FDG.RID==total.loc[row].RID].NORM_MEAN.values[0]
        
    if total.loc[row].VOLUMES==1:
        #'N_HIPPOCAMPUS_VOLUME', 'N_AMYGDALA_VOLUME','N_CTX_ENTORHINAL_VOLUME','N_CTX_CAUDALMIDDLEFRONTAL_VOLUME'
        total.loc[row,'N_HIPPOCAMPUS_VOLUME']=((VOLUMES[VOLUMES.RID==total.loc[row].RID]['Left-Hippocampus'].values[0]  +VOLUMES[VOLUMES.RID==total.loc[row].RID]['Right-Hippocampus'].values[0]  )/2)/VOLUMES[VOLUMES.RID==total.loc[row].RID]['EstimatedTotalIntraCranialVol'].values[0]
        total.loc[row,'N_AMYGDALA_VOLUME']=((VOLUMES[VOLUMES.RID==total.loc[row].RID]['Left-Amygdala'].values[0]  +VOLUMES[VOLUMES.RID==total.loc[row].RID]['Right-Amygdala'].values[0]   )/2) /VOLUMES[VOLUMES.RID==total.loc[row].RID]['EstimatedTotalIntraCranialVol'].values[0]
        total.loc[row,'N_CTX_ENTORHINAL_VOLUME']=((VOLUMES[VOLUMES.RID==total.loc[row].RID]['lh-entorhinal'].values[0]  +VOLUMES[VOLUMES.RID==total.loc[row].RID]['rh-entorhinal'].values[0]   )/2)/ VOLUMES[VOLUMES.RID==total.loc[row].RID]['EstimatedTotalIntraCranialVol'].values[0]
        total.loc[row,'N_CTX_CAUDALMIDDLEFRONTAL_VOLUME']=((VOLUMES[VOLUMES.RID==total.loc[row].RID]['lh-caudalmiddlefrontal'].values[0]  +VOLUMES[VOLUMES.RID==total.loc[row].RID]['rh-caudalmiddlefrontal'].values[0]   )/2)/VOLUMES[VOLUMES.RID==total.loc[row].RID]['EstimatedTotalIntraCranialVol'].values[0]
        
        
        
        
```
